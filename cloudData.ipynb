```python
def replaceinFile(fileName):
    #read input file
    fin = open(fileName, "rt")
    #read file contents to string
    data = fin.read()
    #replace all occurrences of the required string
    data1=data.replace(';', ',')
    data1=data1.replace('\t','')
    #close the input file
    fin.close()
    #open the input file in write mode
    fin = open(fileName, "wt")
    #overrite the input file with the resulting data
    fin.write(data1)
    #close the file
    fin.close()
```

```python
allCsv=[]
allFolders=['2013-7','2013-8','2013-9']
for nF in allFolders:
    for root, dirs, files in os.walk(r'C:/Users/ajeya/Time-Series-Forecasting-/rnd/'+nF+'/'):
        # select file name
        for file in files:
            # check the extension of files
            if file.endswith('.csv'):
                # print whole path of files
                allCsv.append(os.path.join(root, file))

```

```python
allCsv
```

```python
for filePath in allCsv:
    replaceinFile(filePath)
```

```python
def checkTable(table):
    print("============Table Data==========")
    print("\tNa Values\n")
    print(table.isna().sum())
    print("\n\tZeroes Values\n")
    for col in table.columns:
        print(f'Column:{col}---------\nNumber Of Zeroes:{(table[col]==0).sum()}')
    print("\n\tNul
          l Values\n")
    for col in table.columns:
        print(f'Column:{col}---------\nNumber Of Zeroes:{table[col].isnull().sum()}')
```
