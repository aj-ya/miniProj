```python
allVal=[]
for indC in allCsv:
    tempDf=pd.read_csv(indC)
    allVal.append(tempDf['CPU usage [MHZ]'].tolist())
```

```python
allVal[0]
```

```python
open1=[]
min1=[]
max1=[]
close1=[]
for ind in allVal:
    open1.append(ind[0])
    min1.append(min( ind ) )
    max1.append(max(ind))
    close1.append(ind[-1])
```

```python
print(open1[:2],allVal[0][0],allVal[1][0])
print(close1[0:2],allVal[0][-1],allVal[1][-1])
print(min1[0:2],min(allVal[0]),min(allVal[1]))
print(max1[0:2],max(allVal[0]),max(allVal[1]))
```

```python
finaldf = pd.DataFrame(list(zip(open1,min1,max1,close1,allVal)),
               columns =['open', 'min','max','close','all values'])
```

```python
finaldf
```

```python
minOfMax=0     #min(finaldf['max'])
zeroRows=[]
for ind in finaldf.index:
    if(finaldf['max'][ind]==minOfMax):
        zeroRows.append(ind)
```

```python
normLow=[]
normHigh=[]
thresholdMin=20 #lesser than or equal to
thresholdMax=80 #greater than or equal to
```

```python
for ind in finaldf.index:
    k=thresholdMin
    normLow.append((finaldf['max'][ind]*(k/100))+finaldf['min'][ind])
    k=thresholdMax
    normHigh.append((finaldf['max'][ind]*(k/100))+finaldf['min'][ind])
```

```python
print(normLow[0:5])
print(normHigh[0:5])
```

```python
finaldf['Rel-20%']=normLow
finaldf['Rel-80%']=normHigh
```

```python
provhigh=[]
provlow=[]
```

```python
allProvVal=[]
for indC in allCsv:
    tempDf=pd.read_csv(indC)
    allProvVal.append(tempDf['CPU capacity provisioned [MHZ]'].mean())
    
for ind in allProvVal:
    k=thresholdMin
    provlow.append(ind*k/100)
    k=thresholdMax
    provhigh.append(ind*k/100)
```

```python
finaldf['Prov-20%']=provlow
finaldf['Prov-80%']=provhigh
finaldf['CPU capacity provisioned [MHZ]']=allProvVal
```

```python
def checkTable(table):
    print("============Table Data==========")
    print("\tNa Values\n")
    print(table.isna().sum())
    print("\n\tZeroes Values\n")
    for col in table.columns:
        print(f'\nColumn:{col}---------\nNumber Of Zeroes:{(table[col]==0).sum()}')
    print("\n\tNull Values\n")
    for col in table.columns:
        print(f'\nColumn:{col}---------\nNumber Of Zeroes:{table[col].isnull().sum()}')
```
